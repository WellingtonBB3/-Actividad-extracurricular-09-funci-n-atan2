**Investigue y resuma la función atan2**  
La función `atan2` se encarga de calcular el arcotangente de dos números, representados como `x` e `y`, coordenadas de un punto. Esta función devuelve la tangente inversa de cuatro cuadrantes en un rango de -π a π.  
**Restricciones**  

1. Cuando `x=0`, el cálculo de `y/x` en la función `atan` genera una división por cero. Sin embargo, `atan2` maneja este caso correctamente:
- Si `y>0` y `x=0`, devuelve `π/2` (90°).
- Si `y<0` y `x=0`, devuelve `-π/2` (-90°).  
  
2. `atan2(0,0)` devuelve `0` (aunque algunos sistemas pueden considerarlo indefinido).

3. `atan2(y,0)` devuelve `π/2` o `-π/2` dependiendo del signo de `y`.

4. `atan2(0,x)` devuelve `0` o `π` dependiendo del signo de `x`.
  
  **Utilidad**  
  Se utiliza principalmente cuando se requiere trabajar con coordenadas cartesianas, pues calcula el angulo directamente que forma el vector de los puntos (x,y), sin requerir de conversionaes manuales de las coordenadas.

**¿Por qué se recomienda usar la función atan2?**  
Se recomienda utilizar la funcion atan2, pues onsidera el cuadrante del ángulo, ya que a diferencia de `atan`, que solo calcula el ángulo en el rango (-π/2, π/2) (primer y cuarto cuadrante), `atan2` devuelve valores en el rango (-π, π), permitiendo distinguir correctamente los cuadrantes. Ademas evita indeterminaciones, ya que si solo se usa `atan(y/x)`, puede haber problemas cuando `x` es cero (división por cero). `atan2` maneja estos casos correctamente. Por ultimo es más robusta porque se utiliza frecuentemente en gráficos, cálculos de direcciones y aplicaciones de física o ingeniería, donde la orientación en un espacio 2D importa.

**Diferencias entre función atan y atan2.**  

| Aspecto            | atan                                      | atan2                                      |
|--------------------|-------------------------------------------|--------------------------------------------|
| Parámetros         | Toma un solo argumento (tanθ = y/x).       | Toma dos argumentos (y y x).               |
| Rango de salida    | (-π/2, π/2) (ángulos de -90° a 90°).       | (-π, π) (ángulos de -180° a 180°).         |
| Cuadrantes         | No distingue el cuadrante del ángulo.      | Identifica correctamente el cuadrante.     |
| Manejo de x = 0    | Puede generar errores o indeterminación.   | Maneja correctamente casos como x = 0.     |
| Uso en gráficos    | Menos común en gráficos 2D.                | Frecuentemente usado en gráficos 2D.       |
| Robustez           | Menos robusta en cálculos de dirección.    | Más robusta en cálculos de dirección.      |

```python
import math
import matplotlib.pyplot as plt

# Ejemplo 1: Calcular el ángulo entre dos puntos en el plano cartesiano
def calcular_angulo(x1, y1, x2, y2):
    delta_x = x2 - x1
    delta_y = y2 - y1
    angulo = math.atan2(delta_y, delta_x)
    return angulo, delta_x, delta_y

# Puntos (1, 2) y (4, 6)
x1, y1 = 1, 2
x2, y2 = 4, 6
angulo, delta_x, delta_y = calcular_angulo(x1, y1, x2, y2)
print(f"El ángulo entre los puntos (1, 2) y (4, 6) es: {math.degrees(angulo)} grados")

# Graficar los puntos y el vector resultante
plt.figure()
plt.plot([x1, x2], [y1, y2], marker='o')
plt.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right')
plt.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='right')
plt.quiver(x1, y1, delta_x, delta_y, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(0, 5)
plt.ylim(0, 7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ángulo entre dos puntos')
plt.grid()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def direccion_vector(x, y):
    """Calcula la dirección del vector en radianes."""
    angulo = math.atan2(y, x)
    return angulo

# Vector (7, -3)
x, y = 7, -3
direccion = direccion_vector(x, y)
print(f"La dirección del vector ({x}, {y}) es: {math.degrees(direccion):.2f} grados")

# Graficar el vector
plt.figure()
plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='r', label=f"Dirección: {math.degrees(direccion):.2f}°")

# Configurar los límites del gráfico
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Línea horizontal
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Línea vertical
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dirección de un vector en el plano 2D')
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas de los puntos (vectores)
vectores = [(1, 1), (-1, 1), (-1, -1), (1, -1), (2, 3), (-3, 2), (0, 4), (4, 0)]

# Calculamos los ángulos usando atan2
angulos = [math.atan2(y, x) for x, y in vectores]

# Convertimos los ángulos a grados
angulos_grados = [math.degrees(angulo) for angulo in angulos]

# Gráfica
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Dibujamos los vectores y sus ángulos
for (x, y), angulo in zip(vectores, angulos_grados):
    plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='blue')
    plt.text(x * 1.1, y * 1.1, f'{round(angulo, 2)}°', color='red')

# Configuración de la gráfica
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.title('Direcciones de vectores y sus ángulos (grados)')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas del círculo
radio = 5
num_puntos = 50
theta = np.linspace(0, 2 * np.pi, num_puntos)
x = radio * np.cos(theta)
y = radio * np.sin(theta)

# Calculamos los ángulos usando atan2
angulos = [math.atan2(yi, xi) for xi, yi in zip(x, y)]

# Convertimos los ángulos a grados
angulos_grados = [math.degrees(a) for a in angulos]

# Gráfica del círculo
plt.figure(figsize=(8, 8))
plt.plot(x, y, label='Trayectoria circular')
plt.scatter(0, 0, color='red', label='Centro (0,0)')

# Dibujamos líneas radiales desde el centro al círculo
for xi, yi, angulo in zip(x, y, angulos_grados):
    plt.plot([0, xi], [0, yi], color='gray', linestyle='--')
    plt.text(xi * 1.1, yi * 1.1, f'{round(angulo, 2)}°', fontsize=8)

# Configuración de la gráfica
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Ángulos de un movimiento circular')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Puntos que definen la ruta (x, y)
ruta = [(0, 0), (2, 3), (4, 5), (6, 3), (8, 6), (10, 8)]

# Separar las coordenadas X e Y
x_coords = [p[0] for p in ruta]
y_coords = [p[1] for p in ruta]

# Cálculo de los ángulos entre puntos consecutivos
angulos = []
for i in range(len(ruta) - 1):
    x1, y1 = ruta[i]
    x2, y2 = ruta[i + 1]
    angulo = math.atan2(y2 - y1, x2 - x1)
    angulos.append(angulo)

# Convertimos los ángulos a grados
angulos_grados = [math.degrees(a) for a in angulos]

# Gráfica de la ruta
plt.figure(figsize=(10, 6))
plt.plot(x_coords, y_coords, marker='o', label='Ruta', color='blue')

# Dibujar las flechas de dirección
for i in range(len(ruta) - 1):
    x1, y1 = ruta[i]
    x2, y2 = ruta[i + 1]
    dx, dy = x2 - x1, y2 - y1
    plt.quiver(x1, y1, dx, dy, angles='xy', scale_units='xy', scale=1, color='red', alpha=0.7)
    plt.text((x1 + x2) / 2, (y1 + y2) / 2, f'{round(angulos_grados[i], 1)}°', fontsize=8, color='green')

# Configuración de la gráfica
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Ruta con direcciones angulares entre puntos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Parámetros iniciales
posicion_radar = (0, 0)  # El radar está en el origen
trayectoria_objetivo = [(2, 3), (4, 5), (6, 3), (8, 6), (10, 8), (12, 5)]  # Puntos del objetivo
x_radar, y_radar = posicion_radar

# Gráfica del radar y trayectoria del objetivo
plt.figure(figsize=(10, 6))

# Dibujar la trayectoria del objetivo
x_objetivo = [p[0] for p in trayectoria_objetivo]
y_objetivo = [p[1] for p in trayectoria_objetivo]
plt.plot(x_objetivo, y_objetivo, label='Trayectoria del objetivo', color='blue', linestyle='--')
plt.scatter(x_objetivo, y_objetivo, color='green', label='Puntos del objetivo')

# Dibujar el radar
plt.scatter(x_radar, y_radar, color='red', label='Radar (Fijo)')

# Calcular los ángulos de orientación del radar hacia cada punto
angulos = []
for punto in trayectoria_objetivo:
    x_obj, y_obj = punto
    angulo = math.atan2(y_obj - y_radar, x_obj - x_radar)  # Ángulo en radianes
    angulos.append(angulo)
    # Dibujar la línea de seguimiento del radar
    plt.quiver(x_radar, y_radar, x_obj - x_radar, y_obj - y_radar, angles='xy', scale_units='xy', scale=1, color='orange', alpha=0.7)

    # Mostrar el ángulo en grados sobre la línea
    angulo_grados = math.degrees(angulo)
    plt.text((x_obj + x_radar) / 2, (y_obj + y_radar) / 2, f"{round(angulo_grados, 1)}°", fontsize=8, color='purple')

# Configuración de la gráfica
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Simulación de radar rastreando un objetivo')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.xlim(-1, max(x_objetivo) + 2)
plt.ylim(-1, max(y_objetivo) + 2)
plt.show()
```
